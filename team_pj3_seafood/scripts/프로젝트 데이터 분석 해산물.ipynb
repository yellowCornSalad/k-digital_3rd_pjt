## pdf1, pdf2 데이터 통합

```python
pdf = pd.concat([pdf1, pdf2], axis=0)
```

```python
pdf.info()
```

```python
pdf.columns
```

```python
pdf.columns = ['date', 'species', 'origin', 'unit', 'amount', 'weight', 'avg_price']
```

```python
sea_pdf = pdf.copy()
```

```python
mask = (sea_pdf.species == '(활)넙치')|(sea_pdf.species == '(활)왕게')|(sea_pdf.species == '(선)갑오징어')|(sea_pdf.species == '(선)은갈치')|(sea_pdf.species == '(선)참돔')|(sea_pdf.species == '(활)암꽃게')|(sea_pdf.species == '(활)전복')|(sea_pdf.species == '(선)고등어')|(sea_pdf.species == '(활)우럭')|(sea_pdf.species == '(선)가자미')


sea_pdf2 = sea_pdf.loc[mask, :]
```

```python
sea_pdf2.head()
```

```python
sea_pdf2.info()
```

```python
sea_pdf2.to_csv('cleansing_data.csv', index=True)
```

```python
groupby_cols = ['species']

for groupby_col in groupby_cols:
    print('------ ' + groupby_col + ' value count ------')
    sdf2.groupBy(groupby_col).count().show()
```
